```python
contents.pop(len(contents)-1)

output = pd.DataFrame()

for item in contents:
    dict1={}

    # extract sentence id from text
    sender = re.search(r"Sentence id.*", item)

    # extract digit from sentence id
    if sender is not None:
        s_id = re.search(r"\d+", sender.group())
    else:
        s_id=None

    #converting match object into string
    if s_id is not None:
        sentence_id = s_id.group()
    else:
        sentence_id = None

    # Add sentence id to dictionary.
    dict1["sentence_id"] = int(sentence_id)-1

    #code for extracting chunk id
    list2=[]
    sender1 = regex.findall(r"\d+\.\d", item)
    for i in sender1:
        s=regex.search(r"\d+", i)
        if s is not None:
            chun_label = s.group()
        else:
            chun_label = None
        list2.append(chun_label)
    # Add Chunk id to dictionary.
    dict1["chunk_id"]=list2

    #code for extracting chunk label
    list3=[]
    sender2 = regex.findall(r"\(\(\t\w+\t", item)
    sender2
    for i in sender2:
        s2=regex.search(r"\w+", i)
        if s2 is not None:
            chun_label2 = s2.group()
        else:
            chun_label2 = None
        list3.append(chun_label2)
    for i in range(0,len(list2)-1):
        if list2[i+1]==list2[i]:
            list3.insert(i+1, list3[i])
    # Add Chunk id to dictionary.
    dict1["chunk_label"]=list3

    #code for extracting chunk feature set
    list4=[]
    feature_set = regex.findall(r"\d+\t\(\(.*", item)
    for i in feature_set:
        feat_set = regex.search(r"<fs\s.*", i)
        if feat_set is not None:
            chunk_feature_set = feat_set.group()
        else:
            chunk_feature_set = None
        op = chunk_feature_set.replace(" af=',,,,,,,' ", " ")
        list4.append(op)
    for i in range(0,len(list2)-1):
        if list2[i+1]==list2[i]:
            list4.insert(i+1, list4[i])
    # Add Chunk feature set to dictionary.
    dict1["chunk_feature_set"]=list4


    # code for extreacting word id
    sender5 = regex.findall(r"\d+\.\d", item)
    # Add word id to dictionary.
    dict1["word_id"]=sender5

    #code for extracting Malayalam words and position
    word=[]
    pos=[]

    feat_set = regex.findall(r"\d+\.\d+\s.*", item)
    for i in feat_set:
        word.append(i.split()[1])
        pos.append(i.split()[2])

    # add word to dict1
    dict1['word']=word

    #add pos to dict1
    dict1['Pos']=pos

    #code for word feature set
    list8=[]
    feature_set = regex.findall(r"\d+\.\d+\s.*", item)
    for i in feature_set:
        feat_set = regex.search(r"<fs\s.*", i)
        if feat_set is not None:
            chunk_feature_set = feat_set.group()
        else:
            chunk_feature_set = None
        list8.append(chunk_feature_set)
    # Add word feature to dictionary.
    dict1["word_feature_set"]=list8

    # Create dataframe from the dictionary
    samp = pd.DataFrame.from_dict(dict1)

    # append method
    output = pd.concat([output, samp], ignore_index=True)

# save the new dataframe in .csv file
output.to_csv('MalText50.csv')
```

```python
'''
To convert chunk_feature_set to dictionary values
'''
import re

def convert_fs_string_to_dict(fs_string):
 if not fs_string:
   return {}

 pattern = r"<fs\s+(.*?)\s*>"
 match = re.match(pattern, fs_string)
 if match:
   attributes_str = match.group(1)
   attributes = {}
   for pair in attributes_str.split(" "):
     if pair:
       key, value = pair.split("=")
       attributes[key.strip()] = value.strip("'")
   return attributes
 else:
   return {}
```

```python
output['chunk_features_dict'] = output['chunk_feature_set'].apply(convert_fs_string_to_dict)
```

```python
def update_drel(row):
    if 'chunk_features_dict' in row and 'drel' in row['chunk_features_dict']:
        drel_value = row['chunk_features_dict']['drel']
        karaka_r, base = drel_value.split(':')
        updated_drel_value = {'karaka_r': karaka_r, 'base': base}
        row['chunk_features_dict']['drel'] = updated_drel_value
    return row
```

```python
def process_sentence(df, sentence_id):
    """
    Process a single sentence based on its sentence_id.
    """
    # Filter dataframe for the chosen sentence_id
    sentence_df = df[df['sentence_id'] == sentence_id].copy()

    # Reset index to avoid issues with loc
    sentence_df.reset_index(drop=True, inplace=True)

    # Initialize new columns
    sentence_df['base_chunk_id'] = None
    sentence_df['karaka_relation'] = None

    # Find the root chunk for the chosen sentence
    root_chunk = sentence_df[sentence_df['is_root_chunk'] == 'yes']
    root_karaka_relation = None
    if not root_chunk.empty:
        root_chunk_features = root_chunk['chunk_features_dict'].iloc[0]
        root_karaka_relation = root_chunk_features.get('drel', {}).get('karaka_r')

    # Update the root chunk's karaka_relation in the dataframe
    if not root_chunk.empty:
        root_index = root_chunk.index[0]
        sentence_df.loc[root_index, 'karaka_relation'] = root_karaka_relation

    # Iterate through the remaining chunks to find base_chunk_id and karaka_relation
    for index, row in sentence_df.iterrows():
        if row['is_root_chunk'] == 'yes':
            continue
        chunk_features = row['chunk_features_dict']
        base = None
        if 'base' in chunk_features.get('drel', {}):
            base_name = chunk_features['drel']['base']
            base_chunk = sentence_df[sentence_df['chunk_features_dict'].apply(lambda x: x.get('name')) == base_name]
            if not base_chunk.empty:
                base = base_chunk['chunk_id'].iloc[0]
        karaka_relation = chunk_features.get('drel', {}).get('karaka_r')
        sentence_df.loc[index, 'base_chunk_id'] = base
        sentence_df.loc[index, 'karaka_relation'] = karaka_relation

    return sentence_df

def process_dataframe(df):
    """
    Process a dataframe containing multiple sentence_ids.
    """
    # Get unique sentence_ids
    unique_sentence_ids = df['sentence_id'].unique()

    # Process each sentence_id and store the results in a list
    processed_dfs = []
    for sentence_id in unique_sentence_ids:
        processed_dfs.append(process_sentence(df, sentence_id))

    # Combine the processed dataframes into a single dataframe
    combined_df = pd.concat(processed_dfs, ignore_index=True)

    return combined_df

# Example usage:
# combined_df = process_dataframe(df)
```
